```python
# nominal column encoding using one hot encoding
from sklearn.preprocessing import OneHotEncoder

encode_ = ['gender', 'marital_status']

ohe = OneHotEncoder(sparse_output=False)
encoded_data = ohe.fit_transform(sample[encode_])
encoded_ = pd.DataFrame(encoded_data, columns=ohe.get_feature_names_out(encode_))
sample = pd.concat([sample, encoded_], axis=1)
sample.drop(columns=encode_, inplace=True)
```
